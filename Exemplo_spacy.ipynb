```python
import pandas as pd
import numpy as np
df = pd.read_csv('covid19_tweets.csv')
df.head(10)
```

```python
df.shape
```

```python
from pathlib import Path
nlp = en_core_web_sm.load()
```

```python
import random
random.seed(44)
texto = df.text[random.sample(range(1,10000),1000)]
texto
```

```python
combinar_textos = str(texto)
doc = nlp(combinar_textos)
```

```python
for ent in doc.ents:
    print(ent.text, '->', ent.label_)
```

```python
spacy.displacy.render(doc, style ='ent', jupyter = True)

# podemos obserbar que a biblioteca americana j√° tem mais recursos treinados como: dinheiro, organiza√ß√£o,cardinal, local
```

```python
print("GPE:", spacy.explain('GPE'))
print("MONEY:", spacy.explain('MONEY'))
```
